**Question 1**

```r
river = read.table("rivers.txt", header = FALSE)
head(river)
```

```r
# Convert to vector since there is only one column 
river = river[,1]
str(river)
```

**Question 2**

```r
par(mfrow = c(1, 2))
hist(river)
boxplot(river)
```

**Question 3**

```r
qqnorm(river, col = 'blue')
qqline(river, col = 'red')
```



*   From the histogram, the data is right skewed and not normally distributed. The qqplot confirms this since the points deviate from the straight line. 



**Question 4**

```r
sample_rivers <- numeric(1000)
for (i in 1:1000){sample_rivers[i] <- mean(sample(rivers, 5, replace=FALSE))}
```

```r
hist(sample_rivers)
```



*   The histogram is right skewed and hence, the sample is not normally distributed.



**Question 5**

```r
for (i in 1:1000){sample_rivers[i] <- mean(sample(rivers, 30, replace=FALSE))}
```

```r
hist(sample_rivers)
```

```r
qqnorm(sample_rivers)
qqline(sample_rivers)
```



1.   The data is more normally distributed and less skewed.
