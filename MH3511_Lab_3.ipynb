**Question 1**

```r
# Read the file 
wip = read.csv('wip.txt', sep = ' ')
head(wip)
```

```r
str(wip); 
summary(wip)
```

**Question 2**

```r
# Seperate the data according to plants 

wip_1 = wip[wip$plant == 1, ]
wip_2 = wip[wip$plant == 2, ]
```

```r
mean(wip_1$time); 
median(wip_1$time); 
quantile(wip_1$time, 0.25); 
quantile(wip_1$time, 0.75); 
min(wip_1$time); 
max(wip_1$time); 
range(wip_1$time); 
IQR(wip_1$time); 
var(wip_1$time); 
sd(wip_1$time)
```

```r
mean(wip_2$time); 
median(wip_2$time); 
quantile(wip_2$time, 0.25); 
quantile(wip_2$time, 0.75); 
min(wip_2$time); 
max(wip_2$time); 
range(wip_2$time); 
IQR(wip_2$time); 
var(wip_2$time); 
sd(wip_2$time)
```

**Question 3**

```r
par(mfrow = c(1,2))

hist(wip_1$time, 
xlab = "Time", 
main = "Plant 1")

hist(wip_2$time, 
xlab = "Time", 
main = "Plant 2")
```

Both the processing time for plant 1 and 2 are skewed to the right.

```r
par(mfrow = c(1,2))

boxplot(wip_1$time, xlab = "Time", main = "Plant 1")
boxplot(wip_2$time, xlab = "Time", main = "Plant 2")
```

The processing time for plant 2 has a higher median than that of plant 1. The processing time of both plants has a an outlier at the right tail. 

**Question 4**

```r
summary(wip_1$time); 
summary(wip_2$time)
```

The processing time for plant 2 has a higher median. It alos has a higher mean and a larger spread.

**Question 1**

```r
baby = read.table('babies.data', header = TRUE)
head(baby)
```

```r
str(baby); 
summary(baby)
```

**Question 2**

```r
new_baby = baby[baby$smoke != 9, ]
```

**Question 3**

```r
no_smoke = new_baby[new_baby$smoke == 0, ]
smoke = new_baby[new_baby$smoke == 1, ]

str(no_smoke); # 742
str(smoke) # 484 
```

**Question 4**

```r
summary(smoke); 
summary(no_smoke)
```

Babies whose mother smoked during pregnancy have a lower mean and median weight. 

**Question 5**

```r
par(mfrow = c(1,2))

hist(smoke$bwt, xlab = "Weight", main = "Mother smoked")
hist(no_smoke$bwt, xlab = "Weight", main = "Mother did not smoke")
```

```r
par(mfrow = c(1,2))
boxplot(smoke$bwt, xlab = "Weight", main = "Mother smoked")
boxplot(no_smoke$bwt, xlab = "Weight", main = "Mother did not smoke")
```

Boxplots give a better comparison of quantiles. 

**Question 6**

```r
par(mfrow = c(1,2))

qqnorm(smoke$bwt)
qqline(smoke$bwt, col = "red")

qqnorm(no_smoke$bwt)
qqline(no_smoke$bwt, col = "red")
```

Weights of babies whose mother did not smoke during pregnancy differ from the normal distribution at the tails. The weights of babies whose mother smoked appear to be normally distributed.

```r
skull = read.table('skull.txt', header = FALSE)
```

```r
stem(skull[,1], scale = 0.5)
```

121 is an outlier.

**Question 2**

```r
boxplot(skull[, 1]); 
IQR(skull[,1])
```

**Question 3**

```r
s = skull[,1]

x = seq(min(s), max(s), by = 0.1)
n = dnorm(x, mean(s), sd(s))

hist(s, freq = FALSE, xlab = "Height", main = "Skull")
lines(x, n, col = 'red')
```

**Question 4**

```r
qqnorm(s); 
qqline(s, col = 'red')
```

The data is normally distributed except at the tail. 

**Question 5**

```r
par(mfrow = c(1,2))

qqnorm(sqrt(s), main = "Square root transformation")
qqline(sqrt(s), col = 'red')

qqnorm(log(s), main = "log transformation")
qqline(log(s), col = 'red')
```

The transformations did not make the data more normally distributed.
