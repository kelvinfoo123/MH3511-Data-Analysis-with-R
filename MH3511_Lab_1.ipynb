```r
# Exercise 5.1 
matrix4x2 <- matrix(seq(from = 1, to = 8, by = 1), nrow = 4, ncol = 2)
matrix4x2

# Exercise 5.2 
c(rep(1, 4)) %*% matrix4x2

# Exercise 5.3 
matrix2x4 <- matrix(c(1,2,3,4,11,12,13,14), nrow = 2, ncol = 4, byrow = TRUE)
matrix2x4

# Exercise 5.4 
matrix4x2 %*% matrix2x4 
matrix2x4 %*% matrix4x2

# Exercise 5.5 
matrix2x2 <- matrix4x2[c(1,2), ]
matrix2x2

# Exercise 5.6
# The matrix is invertible 
solve(matrix2x2)

# Exercise 5.7
diag3 <- diag(c(-1, -2, -3))
diag3

# Exercise 5.8 
c(2,4,8) %*% diag3 
c(2,4,8) * c(-1, -2, -3)
```
