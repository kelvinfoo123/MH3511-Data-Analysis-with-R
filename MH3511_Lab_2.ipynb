```r
fixed = read.fwf('fixed.txt', widths = c(3, 1, 3, 2, 1))
head(fixed)
```

```r
str(fixed)
```

```r
# Assign names to columns

colnames(fixed) = c('id', 'gender', 'height', 'weight', 'siblings')
head(fixed)
```

```r
test = read.csv('test.txt', sep = ' ')
head(test)
```

```r
# Merge fixed and test 

merge = merge(fixed, test)
head(merge)
```

```r
# Modify weight of subject 211

merge[merge$id == 211, "weight"] = 77 
subset(merge, id == 211)
```

```r
# Test score of individuals whose heights are greater than 182 

subset(merge, height > 182)
subset(merge, height > 182)$test
```

```r
# Tallest female and obtain her height, weight and test score 

max(merge[merge$gender == 'F', 'height'])

merge[(merge$gender == 'F') & (merge$height == 178), ]
```

```r
euclidean = function(a,b){
  r = a %% b 
  q = (a - r)/b
  return (list(q,r))
}
```

```r
euclidean(27,12)
```

```r
cm4 = function(x){ 
  moment = sum((x - mean(x))**4) / length(x)
  return (moment)
}
```

```r
x = c(1, 2, 4, 6, 2, 1)
cm4(x)
```
